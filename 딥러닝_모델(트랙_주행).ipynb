**# tensorflow 라이브러리**

딥러닝 모델을 구축하고 훈련시키기 위한 오픈 소스 라이브러리

* Sequential : Keras의 모델 중 하나로 레이어를 순차적으로 쌓아 신경망을 구축하는데 사용 ('Nvidia_Model' 사용)

    Dense는 밀집 연결 레이어로 완전 연결 신경망 레이어

**데이터 불러오기**

-코드 추가 설명
* os.path.join() : 여러 경로 구성 요소를 연결하여 단일 경로로 변경

    여기서는 data_dir이 '/content/Videos'이고 filename이 'train_00000_090.png' 이므로 '/content/Videos/train_00000_090.png' 와 같은 경로 생성

* DataFrame() : 데이터 프레임은 행과 열로 구성된 테이블 형식의 데이터 구조

    여기서 빈 데이터 프레임을 생성하고 df [ 'ImagePath' ] = image_paths 와 df [ 'Angle' ] = steering_angles 코드를 사용해 데이터 프레임에 이미지 경로와 조향각 정보를 각각 열에 추가

````python
data_dir = '/content/Videos'                                           # video.zip파일을 압축 해제하고 생성된 Videos파일로 폴더 경로를 지정해줌
file_list = os.listdir(data_dir)                                       # os.listidir() 함수를 사용해 data_dir에 지정된 폴더 내의 파일 목록을 가져옴
image_paths = []                                                       # 이미지 파일의 경로를 저장할 빈 리스트 생성
steering_angles = []                                                   # 조향각 정보를 저장할 빈 리스트 생성
pattern = "*.png"                                                      # 파일명에 '.png'가 포함된 파일을 fnmatch모듈로 찾기 위해 'pattern'변수에 넣어줌

for filename in file_list:                                             # file_list에 있는 각 파일 이름에 대해 '*.png'이 있는 파일을 확인
    if fnmatch.fnmatch(filename, pattern):
        image_paths.append(os.path.join(data_dir,filename))            # 생성해뒀던 빈 리스트 'image_paths'에 'data_dir'과 'filename'을 합친 경로를 추가
        angle=int(filename[-7:-4])                                     # 파일 이름의 끝에서 7번째부터 5번째까지 추출하여 'angle' 변수에 저장
        steering_angles.append(angle)                                  # 생성해뒀던 빈 리스트 'steering_angles'에 'angle'값 추가

image_index=100                                                        # 100번째 이미지 경로와 각도, 이미지를 출력해보기 위해 인덱스 지정
plt.imshow(Image.open(image_paths[image_index]))                       # 해당 인덱스의 이미지 경로를 열어 이미지를 시각화
print(" image_path :%s" % image_paths[image_index])                     # 해당 인덱스의 이미지 경로를 출력
print(" steering_Angle: %d" % steering_angles[image_index])             # 해당 인덱스의 조향각 정보를 출력

df = pd.DataFrame()                                                    # 'df'라는 빈 데이터 프레임 생성
df['ImagePath'] = image_paths                                          # 'df'에 'ImagePath'열을 생성하여 이미지경로 추가
df['Angle'] = steering_angles                                          # 'df'에 'Angle'열을 생성하여 조향각 정보 추가
````

**조향각에 따른 이미지 개수 시각화**

히스토그램은 데이터의 분포를 시각화하는데 사용하며 빈(bins)의 수가 많을수록 데이터의 세부적인 분포를 더 잘 보여줌

    빈(bins) : 데이터를 일정한 간격으로 나누는 구간

        조향각은 3가지인데 3이 아닌 25로 하는 이유는 3으로 하면 각도축에서 45-135도를 단순 3등분 해버리기 때문에 실제 값과 차이를 보임
        따라서 25등분 정도가 적정함

Figure : matplotlib에서 그림 전체를 의미, Figure에는 하나 이상의 서브플롯이 존재할 수 있음

    서브플롯 : Figure내에 위치한 개별적인 그래프 영역으로 행과 열로 정의됨

              행이 2이고 열이 2이면 2*2 = 4, 4개의 서브플롯
              여기서는 조향각에 대해서 개수를 표현하는 그래프 이므로 단일 서브플롯

````python
num_of_bins = 30                                                   # 히스토그램을 그릴 때 사용하는 빈(bins)의 개수를 지정하기 위해 생성한 변수
hist, bins = np.histogram(df['Angle'], num_of_bins)                # 데이터프레임에 포함된 'Angle' 열(column) 데이터를 가져와 hist에 할당, bins에 num_of_bins 할당

fig, axes = plt.subplots(1, 1, figsize=(12, 4))                    # 1, 1 이므로 단일 서브플롯 / figsixe = (12, 4)는 크기 설정 - 너비가 12인치, 높이가 4인치
axes.hist(df['Angle'], bins=num_of_bins, width=1, color='blue')    # df['Angle']데이터를 사용해 bins는 num_of_bins, 막대 너비는 1, 색은 blue로 지정해서 히스토그램 생성

angles = [45.4, 90.5, 132.4]

plt.title('Number of Image by angle')
plt.ylabel('image_count')
plt.xlabel('steering_angles')
plt.xticks(angles, ['45(left)', '90(straight)', '135(right)'])
````

정규화를 255로 해주는 이유 : 보통 이미지의 픽셀 값은 0부터 255까지의 범위에 있음. 이를 255로 나눔으로서 0부터 1사이의 값으로 바꿔 좀 더 안정적으로 학습될 수 있도록 해줌

스케일링을 해주면 최적화 알고리즘이 더 잘 작동할 수 있음

0과 1사이로 스케일링을 해주는 이유는 일부 활성화 함수(시그모이드, 소프트맥스 등)은 0과 1사이에서 최적으로 동작함

    데이터 스케일링이 작을수록 최적화 알고리즘이 더 빠르게 수렴할 수 있어 입력 데이터를 작은 범위로 스케일링 하면 학습 속도를 향상시킬 수 있음
    학습의 안전성에 있어서도 작은 스케일의 데이터는 오버피팅을 방지하고 학습 과정을 안정화할 수 있음

        * 오버피팅 : 기계학습에서 발생하는 부요 문제 중 하나로 모델이 훈련 데이터에 대해서는 잘 작동하지만 새로운 데이터에 대해서는 성능이 떨어지는 현상

                   오버피팅이 발생하면 모델이 훈련 데이터의 노이즈나 특정한 패턴을 학습하여 새로운 데이터에 대한 일반화 능력이 저하됨
                   원인으로는 너무 많은 훈련 에포크, 훈련 데이터의 부족, 복잡한 모델 구조 등이 있다.
                   훈련 데이터의 양을 늘리거나 모델을 단순화하거나 드롭 아웃 등을 통해 방지할 수 있다.

                   *Dropout : 모델이 과적합 되는 것을 방지하기 위해 일부 뉴런을 무작위로 비활성화

**# Nvidia 모델**

NVIDIA 모델은 자율주행 차량의 차선 인식, 장애물 회피, 경로 계획 등의 작업에서 높은 성능을 발휘
고성능 GPU와 최적화된 소프트웨어 스택을 통해 실시간 비전 작업을 효율적으로 수행할 수 있으며, 자율주행 차량, 드론, 로보틱스 등 다양한 분야에서 사용


```
NVIDIA 모델은 자율주행 시스템에서 다양한 역할을 수행

1. 차선 인식 : 도로의 차선을 인식하고 차량이 차선 내에서 주행할 수 있도록 도움을 줌

2. 장애물 회피 : 실시간으로 장애물을 감지하고 이를 회피하는 경로를 계획

3. 경로 계획 : 카메라 및 센서 데이터를 바탕으로 차량의 주행 경로를 계획

4. 스티어링 제어 : 모델이 예측한 경로에 따라 차량의 스티어링을 제어하여 안전한 주행을 유지
```




````python
def nvidia_model():
    model = Sequential(name = 'Nvidia_Model')

    model.add(Conv2D(24, (5, 5), strides=(2, 2), input_shape = (66, 200, 3), activation = 'elu'))
    model.add(Conv2D(36, (5, 5), strides=(2, 2), activation = 'elu'))
    model.add(Conv2D(48, (5, 5), strides=(2, 2), activation = 'elu'))
    model.add(Conv2D(64, (3, 3), activation = 'elu'))
    model.add(Dropout(0.2))
    model.add(Conv2D(64, (3, 3), activation = 'elu'))

    model.add(Flatten())
    model.add(Dropout(0.2))
    model.add(Dense(100, activation = 'elu'))
    model.add(Dense(50, activation = 'elu'))
    model.add(Dense(10, activation = 'elu'))

    model.add(Dense(1))

    optimizer=Adam(learning_rate=0.001)
    model.compile(loss = 'mse', optimizer = optimizer)

    return model

model = nvidia_model()
print(model.summary())
````

batch : 머신러닝이나 딥러닝에서 모델을 학습시키기 위해 데이터를 소규모 그룹으로 나누어 처리하는 방법

batch_size : 한번의 학습단계에서 사용되는 데이터 샘플의 수
 * batch_size = 100이라면 한번의 학습단계에서 100개의 데이터 샘플을 사용하여 가중치를 업데이트 함
  
  

```
<배치 크기가 100, 전체 데이터가 7666개, 1 에포크 동안 300번의 학습 단계를 가지는 경우>
300번의 학습 단계 동안 매 단계마다 100개의 데이터 샘플을 무작위로 선택

한 에포크에서 300번의 학습단계 중 첫 번째 단계에서 100개의 랜덤 데이터 샘플을 사용하여 가중치를 업데이트함

그 과정을 300번 반복하는 것이 1 에포크

한 에포크 과정을 10번 반복하여 학습
```
데이터를 증강하여 진행할 수도 있지만 랜덤 추출을 하여 증강 없이 사용

결론적으로 중복을 허용하여 30000개의 데이터 샘플을 사용한 것이 됨.


````python
def image_data_generator(image_paths, steering_angles, batch_size):
    while True:
        batch_images = []
        batch_steering_angles = []

        for i in range(batch_size):
            random_index = random.randint(0, len(image_paths) - 1)
            image_path = image_paths[random_index]
            image = my_imread(image_paths[random_index])
            steering_angle = steering_angles[random_index]

            image = img_preprocess(image)
            batch_images.append(image)
            batch_steering_angles.append(steering_angle)

        yield(np.asarray(batch_images), np.asarray(batch_steering_angles))

ncol = 2
nrow = 2

X_train_batch, y_train_batch = next(image_data_generator(X_train, y_train, nrow))
X_valid_batch, y_valid_batch = next(image_data_generator(X_valid, y_valid, nrow))

fig, axes = plt.subplots(nrow, ncol, figsize = (15,6))
fig.tight_layout()

for i in range(nrow):
    axes[i][0].imshow(X_train_batch[i])
    axes[i][0].set_title("training, angle = %s" % y_train_batch[i])
    axes[i][1].imshow(X_valid_batch[i])
    axes[i][1].set_title("validation, angle = %s" % y_valid_batch[i])
````

````python
from sklearn.metrics import mean_squared_error, r2_score

def summarize_prediction (Y_true, Y_pred):                                   # Y_true(실제 값)과 Y_pred(예측 값)으로 mse(평균 제곱 오차)와 r_squared(결정계수)를 계산하는 함수

    mse=mean_squared_error(Y_true, Y_pred)
    r_squared=r2_score(Y_true, Y_pred)

    print(f'mse={mse:2}')
    print(f'r_squared={r_squared:.2%}')
    print()

def predict_and_summarize(X,Y):                                              # 최종 모델을 불러와 X와 Y로 X에 대한 예측을 수행하고 그 값을 Y_pred에 저장
    model=load_model(f'{model_output_dir}/lane_navigation_final.keras')      # 저장된 Y_pred(예측값)와 Y(실제 값)를 사용해 위 함수(mse와 r_squared)를 계산
    Y_pred=model.predict(X)
    summarize_prediction(Y,Y_pred)
    return Y_pred

n_tests=100
X_test, y_test=next(image_data_generator(X_valid,y_valid,100))               # batch_size = 100에 해당하는 데이터를 각각 X_test(경로), y_test(각도)에 저장

y_pred=predict_and_summarize(X_test,y_test)                                  # y_test(각도)와 X_test(경로)를 바탕으로 y_pred(예측 각도)를 만들고 y_test와 y_pred 비교

n_tests_show=2
fig,axes=plt.subplots(n_tests_show,1,figsize=(10,4*n_tests_show))
for i in range(n_tests_show):
    axes[i].imshow(X_test[i])
    axes[i].set_title(f"actual angle={y_test[i]}, predicted angle={int(y_pred[i])},diff={int(y_pred[i])-y_test[i]}")
````
